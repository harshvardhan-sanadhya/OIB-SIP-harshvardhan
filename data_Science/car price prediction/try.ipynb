```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.ensemble import RandomForestRegressor 
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score
```

```python
data=pd.read_csv(r"C:\Users\Lenovo\Downloads\archive (11)\car data.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
le=LabelEncoder()
data['Car_Name']=le.fit_transform(data['Car_Name'])
data['Fuel_Type']=le.fit_transform(data['Fuel_Type'])
data['Selling_type']=le.fit_transform(data['Selling_type'])
data['Transmission']=le.fit_transform(data['Transmission'])
```

```python
data.info()
```

```python
print("Median \n",data.median)
```

```python
x=data.drop(columns=['Selling_Price'])
y=data['Selling_Price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
print(x_train.shape)
print(x_test.shape)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred_lr=lr.predict(x_test)
print("R2 Score:",r2_score(y_test,y_pred_lr))
```

```python
rr=RandomForestRegressor()
rr.fit(x_train,y_train)
y_pred_rr=rr.predict(x_test)
print("R2 Score Random Forest:",r2_score(y_test,y_pred_rr))
```
